# Importing Libraries

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import seaborn as sns
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

# Importing Datasets

In this file, we are doing K means clustering between the cholera outbreak dataset and GDP dataset.

```python
#loading datasets
outbreak_file_path = '../cholera_outbreak_data/outbreak_data.csv'
gdp_file_path = '../gdp_data/gdp_data.csv'

outbreak_data = pd.read_csv(outbreak_file_path)
gdp_data = pd.read_csv(gdp_file_path)
```

```python
#displaying the first few rows of outbreak data
outbreak_data.head()
```

We have proper cholera outbreak data for the year range of 2010 - 2020 so we will only have that data in the gdp data. 

```python
gdp_data = gdp_data[["Country Name", "Country Code", "2010 [YR2010]", "2011 [YR2011]", "2012 [YR2012]", "2013 [YR2013]", 
          "2014 [YR2014]", "2015 [YR2015]", "2016 [YR2016]", "2017 [YR2017]", "2018 [YR2018]", "2019 [YR2019]", "2020 [YR2020]"]]

#convert year columns to numeric
year_columns = gdp_data.columns[2:]
gdp_data[year_columns] = gdp_data[year_columns].apply(pd.to_numeric, errors='coerce')

#handling null values: fill null values with the mean of the respective column
for col in year_columns:
    gdp_data[col].fillna(0, inplace=True)

#check the first few rows after conversion
gdp_data.head()
```

The cholera data has to modified to allow for easy merging with the GDP dataset. The data has to be grouped by country, which will be each row, and then sum up the cases for each year that there is data available for.

```python
#extracting country codes and years from the data
outbreak_data['year'] = pd.to_datetime(outbreak_data['start_date']).dt.year

#grouping data by country and year, and summing up the total suspected cases
grouped_data = outbreak_data.groupby(['country', 'year'])['total_suspected_cases'].sum().reset_index()

#pivoting the data to have countries as rows and years as columns
pivot_data = grouped_data.pivot(index='country', columns='year', values='total_suspected_cases').reset_index()

#filling NaN values with 0, as no cases can be interpreted as 0 cases
pivot_data.fillna(0, inplace=True)

pivot_data.head()
```

# Merging the Datasets

The cholera outbreak dataset uses country codes while the GDP dataset uses the name of the country so we have to find the unique countries and map the codes to the countries to merge the data.

```python
#checking unique country names in both datasets and their counts
outbreak_countries = pivot_data['country'].unique()
gdp_countries = gdp_data['Country Name'].unique()

outbreak_country_count = len(outbreak_countries)
gdp_country_count = len(gdp_countries)

outbreak_countries, outbreak_country_count, gdp_countries, gdp_country_count
```

```python
country_code_name_mapping = {
    'AGO': 'Angola',
    'BEN': 'Benin',
    'CIV': "Cote d'Ivoire",
    'CMR': 'Cameroon',
    'COD': 'Congo, Dem. Rep.',
    'COG': 'Congo, Rep.',
    'ETH': 'Ethiopia',
    'GHA': 'Ghana',
    'GIN': 'Guinea',
    'GNB': 'Guinea-Bissau',
    'KEN': 'Kenya',
    'MLI': 'Mali',
    'MOZ': 'Mozambique',
    'MWI': 'Malawi',
    'NAM': 'Namibia',
    'NER': 'Niger',
    'NGA': 'Nigeria',
    'SLE': 'Sierra Leone',
    'SSD': 'South Sudan',
    'TCD': 'Chad',
    'TZA': 'Tanzania',
    'TZA_zanzibar': 'Zanzibar',
    'UGA': 'Uganda',
    'ZMB': 'Zambia',
    'SOM': 'Somalia',
    'ZWE': 'Zimbabwe'
}

#mapping the country codes in the outbreak dataset to country names
pivot_data['country_name'] = pivot_data['country'].map(country_code_name_mapping)

#merging the datasets based on the country names
merged_data = pd.merge(
    pivot_data,
    gdp_data,
    left_on='country_name',
    right_on='Country Name',
    how='inner'
)

#displaying the first few rows of the merged dataset for verification
merged_data.head()
```

We need to add mean cholera cases overall and mean gdp for each country.

```python
merged_data["Mean_Cases"] = merged_data.iloc[:, 1:11].mean(axis = 1)
merged_data["Mean_GDP"] = merged_data.iloc[:, 15:].mean(axis = 1)
```

```python
merged_data.head()
```

```python
merged_data.columns = ['Country Code',
                    'cases_2010', 'cases_2011', 
                    'cases_2012', 'cases_2013', 
                    'cases_2014', 'cases_2015', 
                    'cases_2016', 'cases_2017', 
                    'cases_2018', 'cases_2019', 
                    'cases_2020', 
                    'Country Name', 'Country Name GDP', 'Country Code GDP', 
                    'gdp_2010', 'gdp_2011', 'gdp_2012', 'gdp_2013', 'gdp_2014',
                    'gdp_2015', 'gdp_2016', 'gdp_2017', 'gdp_2018', 'gdp_2019', 
                    'gdp_2020', 
                    'MeanCholeraCases', 'MeanGDP']
merged_data = merged_data[np.concatenate([["Country Name", "Country Code", "Country Name GDP", "Country Code GDP", "MeanCholeraCases", "MeanGDP"], 
                merged_data.columns[1:12], merged_data.columns[15:-2]])]
merged_data = merged_data.drop(["Country Name GDP", "Country Code GDP"], axis = 1)
```

```python
merged_data.head()
```

# KMeans

Using the Elbow Plot Method to find the ideal number of clusters.

```python
cases_inertia = []
gdp_inertia = []

for cluster_count in range(1, merged_data.shape[0]):
    
    kmeans = KMeans(cluster_count, init = 'k-means++', random_state = 42)
    kmeans.fit(normalize(merged_data.iloc[:, 4:15]))
    cases_inertia.append(kmeans.inertia_)
    
    kmeans = KMeans(cluster_count, init = 'k-means++', random_state = 42)
    kmeans.fit(normalize(merged_data.iloc[:, 15:]))
    gdp_inertia.append(kmeans.inertia_)
    
plt.figure(figsize = (12, 6))

#plotting the Elbow plot
plt.figure(figsize=(10, 6))
plt.plot(cases_inertia, label = 'Cholera Cases')
plt.title('Elbow Method for Optimal K for Cholera Cases Data')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

#plotting the Elbow plot
plt.figure(figsize=(10, 6))
plt.plot(gdp_inertia, label = 'GDP Data')
plt.title('Elbow Method for Optimal K for GDP Data')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
merged_data.columns
```

Based on the plot, k = 5. 

```python
map_data = merged_data.copy()

kmeans = KMeans(5, init = 'k-means++', random_state = 42)
kmeans.fit(normalize(map_data.iloc[:, 4:15]))
map_data['ClusterForCholeraCases'] = kmeans.labels_
map_data['ClusterForCholeraCases'] = map_data['ClusterForCholeraCases'].astype('category')

kmeans = KMeans(5, init = 'k-means++', random_state = 42)
kmeans.fit(normalize(map_data.iloc[:, 15:-1])) 
map_data['ClusterForGDP'] = kmeans.labels_
map_data['ClusterForGDP'] = map_data['ClusterForGDP'].astype('category')

map_data.head()
```

# Geographic Plots

Prepping the data for the geographic plots

```python
#loading the world map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

#filtering for Sub-Saharan African countries in the world dataset
sub_saharan_africa = world[world['continent'] == 'Africa']

#merging the cholera and GDP data with the geographical data
geo_data = sub_saharan_africa.merge(map_data, how='left', left_on='name', right_on='Country Name')

grouped_cases = geo_data.dropna()[[ "MeanCholeraCases", "ClusterForCholeraCases"]]
grouped_cases = grouped_cases.groupby(["ClusterForCholeraCases"]).mean().reset_index()
grouped_cases
```

```python
# For GDP
grouped_gdp = geo_data.dropna()[["MeanGDP", "ClusterForGDP"]]
grouped_gdp = grouped_gdp.groupby(['ClusterForGDP']).mean().reset_index()
grouped_gdp['MeanGDP'] = grouped_gdp['MeanGDP'].apply(lambda x: '{:.2f}'.format(x))
grouped_gdp
```

```python
#plotting the geographic distribution of Mean Cholera Cases and Mean GDP
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

#cholera Cases Map
geo_data.plot(column='ClusterForCholeraCases', ax=ax[0], legend=True, cmap='Reds', edgecolor = 'black')
ax[0].set_title('Mean Cholera Cases (2010-2020)')
ax[0].axis('off')

#GDP Map
geo_data.plot(column='ClusterForGDP', ax=ax[1], legend=True, cmap='Greens', edgecolor = 'black')
ax[1].set_title('Mean GDP (2010-2020)')
ax[1].axis('off')

plt.show()
```
