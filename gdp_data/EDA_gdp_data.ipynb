```python
import pandas as pd
from google.colab import files
```

```python
# Load the data
data = pd.read_csv('gdp_data.csv')

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
data.shape
data.info()
data.describe
```

```python
# Replace ".." with NaN to handle missing values properly
data.replace("..", pd.NA, inplace=True)

# Convert year columns to numeric
year_columns = data.columns[4:]
data[year_columns] = data[year_columns].apply(pd.to_numeric, errors='coerce')

# Check the first few rows after conversion
data.head()
```

```python
# Converting the year columns to numeric, replacing non-numeric entries with NaN
year_columns = [col for col in data.columns if 'YR' in col]
for col in year_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Check for missing values after conversion
missing_values = data[year_columns].isnull().sum()

missing_values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selecting a few representative years for analysis
selected_years = ['2010 [YR2010]', '2014 [YR2014]', '2018 [YR2018]', '2020 [YR2020]']

# Creating histograms for the selected years
plt.figure(figsize=(20, 15))

for i, year in enumerate(selected_years, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[year].dropna(), kde=False, bins=30)
    plt.title(f'Histogram for {year}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Bar plot for 'Country Name' and 'Series Name'

# Selecting the top 10 most frequent categories for visualization
top_countries = data['Country Name'].value_counts().head(10)
top_series = data['Series Name'].value_counts().head(10)

# Plotting
plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries')
plt.xlabel('Frequency')
plt.ylabel('Country Name')

plt.subplot(2, 1, 2)
sns.barplot(x=top_series.values, y=top_series.index)
plt.title('Top 10 Series')
plt.xlabel('Frequency')
plt.ylabel('Series Name')

plt.tight_layout()
plt.show()
```

```python
# Bar plot for 'Country Name'
plt.figure(figsize=(15, 8))
sns.countplot(y='Country Name', data=data, order = data['Country Name'].value_counts().index)
plt.title('Frequency of Each Country')
plt.xlabel('Count')
plt.ylabel('Country Name')
plt.show()

# Due to potentially large number of unique series, we'll limit the bar plot to the top 20 series
top_series = data['Series Name'].value_counts().head(20)
plt.figure(figsize=(15, 8))
sns.barplot(x=top_series.values, y=top_series.index)
plt.title('Top 20 Series by Frequency')
plt.xlabel('Count')
plt.ylabel('Series Name')
plt.show()

```

```python
# Creating boxplots for the selected years
plt.figure(figsize=(20, 10))

for i, year in enumerate(selected_years, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=data, x=year)
    plt.title(f'Boxplot for {year}')

plt.tight_layout()
plt.show()
```
